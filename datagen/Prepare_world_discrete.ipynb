## Create worlds for assey

```python
%matplotlib inline
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_mldata

#download data
data= fetch_mldata('MNIST original', data_home=".")
```

```python
dpath=os.path.abspath("")
dpath
```

```python
x=data.data
mnist=np.reshape(x,(len(x),28,28))
shuffle = np.random.permutation(len(mnist[:,0,0]))
mnist=np.asarray(mnist[shuffle,:,:])/255.
```

```python
for i in range(9):
    plt.figure()
    plt.imshow(mnist[i],cmap=plt.cm.gray)
```

```python
def create_world_discrete(size=np.array((2,2)), savepath=None):
    digit_matrix = []
    count = 0
    for vertical in range(size[0]):
        data_set = []
        for horizonal in range(size[1]):
            data = mnist[count][np.newaxis,np.newaxis,:,:]
            data_set.append(data)
            #plt.figure()
            #plt.imshow(data, cmap='Greys_r')
            count = count + 1
        mnist_concat = np.concatenate(data_set, axis=1)
        digit_matrix.append(mnist_concat)
    mnist_matrix = np.concatenate(digit_matrix)
    #plt.imshow(mnist_matrix)
    np.save(savepath, mnist_matrix)
    return mnist_matrix
```

## Create world1

```python
hoge=create_world_discrete(size=np.array((2,2)), savepath='mnist_matrix/world1_discrete')
```

```python
plt.imshow(hoge[1,1,:,:])
```

```python
create_world_discrete(size=np.array((3,3)), savepath='mnist_matrix/world2_discrete')
```
