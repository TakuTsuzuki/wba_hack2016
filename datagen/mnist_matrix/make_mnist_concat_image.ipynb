```python
import numpy as np
import matplotlib.pyplot as plt
import chainer
%matplotlib inline
```

```python
train, test = chainer.datasets.get_mnist()
```

```python
train_iter = chainer.iterators.SerialIterator(train, 100)
```

```python
mnist = train_iter.next()
```

```python
plt.imshow(mnist[3][0].reshape([28,28]), cmap='Greys_r')
```

```python
t = []
for i,d in enumerate(mnist):
    t.append(d[1])
t = np.array(t)
```

```python
digit_matrix = []
for num in range(5):
    digit_data_set = []
    for digit in range(10):
        ind = np.linspace(0,99, 100)
        use_index = int(ind[t==digit][num].tolist())
        digit_data = mnist[use_index][0].reshape([28,28])
        digit_data_set.append(digit_data)
        plt.figure()
        plt.imshow(digit_data, cmap='Greys_r')
    mnist_concat = np.concatenate(digit_data_set, axis=1)
    digit_matrix.append(mnist_concat)
```

```python
mnist_matrix = np.concatenate(digit_matrix)
```

```python
plt.imshow(mnist_matrix, 'Greys_r')
plt.savefig('mnist_matrix.png')
```

```python
np.save('mnist_matrix', mnist_matrix)
```
